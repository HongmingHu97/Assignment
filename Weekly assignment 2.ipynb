# Linear regression

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

I am going to find the Linear regression about the number of content and the number of pictures

```python
y= df["num_imgs"]
x= df[["n_tokens_content"]]
lm = LinearRegression()
lm.fit(x,y)
b0 = lm.intercept_
b1 = lm.coef_[0]
print(f'The regression line is equal to y = {b0:.2f} + {b1:.2f}X')
```

The regression line is equal to y= 1.16 + 0.01x

```python
df['num_imgs_p'] = lm.predict(x)
df.head(50)
```

```python
df["resid"] = df["num_imgs"] - df["num_imgs_p"]
df.head()
```

# Evaluating the model

```python
residual = df['num_imgs_p'] - df['num_imgs']
sns.scatterplot(x='num_imgs',y='num_imgs_p',data=df)
plt.xlim(0, 1400000) 
plt.ylim(0, 1400000)
plt.plot([0, 1.5e6], [0, 1.5e6], color='orange', lw=3) 
plt.xlabel('Number of images')
plt.ylabel('Number of images (predicted)')
plt.show()
```

```python
df['residual'] = df['num_imgs_p'] - df['num_imgs']
sns.scatterplot(x='n_tokens_content',y='residual',data=df)
plt.xlabel('The number of words in the content')
plt.ylabel('Residuals')
plt.show()
```

```python
r2_score(df['num_imgs'], df['num_imgs_p'])
```

```python
math.sqrt(mean_squared_error(df['num_imgs'], df['num_imgs_p']))
```

Based on the R^2, We can predict 13% of the variation in the number of pictures based on the number of words in the content.
